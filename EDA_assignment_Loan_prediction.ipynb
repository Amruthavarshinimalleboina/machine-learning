```python
q1=np.quantile(data,0.25)
q2=round(np.quantile(data,0.50),2)
q3=round(np.quantile(data,0.75),2)
IQR=(q3-q1)
LB=q1-1.5*IQR
UB=q3+1.5*IQR
    
print("q1:",q1)
print("q2:",q2)
print("q3:",q3)
print("IQR:",IQR)
print("LB:",LB)
print("UB:",UB)
```

```python
data=loan_prediction['ApplicantIncome']
con1=data<LB
con2=data>UB
con1|con2
```

```python
loan_prediction[con1|con2]
loan_prediction[con1
               &
               con2]
```
